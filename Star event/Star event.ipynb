```python
text = open("input2.txt","r")
lines = text.readlines()
tableA = {"Y":6,"X":3,"Z":0}
tableA_r = {"X":"Z","Y":"X","Z":"Y"}
tableB = {"Z":6,"Y":3,"X":0}
tableB_r = {"X":"X","Y":"Y","Z":"Z"}
tableC = {"X":6,"Z":3,"Y":0}
tableC_r = {"X":"Y","Y":"Z","Z":"X"}
points = {"X":1,"Y":2,"Z":3}
score = 0

for line in lines:
    letters = line.split()
    match letters[0]:
        case "A":
            letter = tableA_r[letters[1]]
            score += tableA[letter] + points[letter]
        case "B":
            letter = tableB_r[letters[1]]
            score += tableB[letter] + points[letter]
        case "C":
            letter = tableC_r[letters[1]]
            score += tableC[letter] + points[letter]

print(score)
```

```python
text = open("input3.txt","r")
lines = text.readlines()
score = 0
letters_both = []
for line in lines:
    sac1 = line[:len(line)//2]
    sac2 = line[len(line)//2:]
    found = False
    for letter in sac2:
        if not found and letter in sac1:
            found = True
            letters_both.append(letter)
    
for letter in letters_both:
    if ord(letter) >= 97:
        score += ord(letter) - 96
    else:
        score += ord(letter) - 38

print(score)
```

```python
text = open("input7.txt","r")
lines = text.readlines()

#Construction de l'arborescence
father = {}
contains = {}
currentDirectory = ""
for line in lines:
    if line[0] == "$":
        if line[2:4] == "cd":
            match line[5:-1]:
                case "..":
                    currentDirectory = father[currentDirectory]
                case "/":
                    currentDirectory = "/"
                    contains[currentDirectory] = []
                case _:
                    father[line[5:-1]] = currentDirectory
                    if not currentDirectory in contains.keys():
                        contains[currentDirectory] = []
                    else:
                        if not "dir "+line[5:-1] in contains.values():
                            contains[currentDirectory].append("dir "+line[5:-1])
                    currentDirectory = line[5:-1]
                    contains[currentDirectory] = []

    else:
        if not line in contains[currentDirectory]:
            contains[currentDirectory].append(line[:-1])

#Calcul de la taille des fichiers
size_dirs = {}
def size_dir(directory):
    size_dirs[directory] = 0
    for line in contains[directory]:
        if line[:3] == "dir":
            size_dir(line[4:])
            size_dirs[directory] += size_dirs[line[4:]]
        else:
            size_dirs[directory] += int(line.split()[0])
    
            
size_dir("/")
score = 0
for size in size_dirs.values():
    if size <= 100000:
        score += size
        
print(score)
print(len(size_dirs.keys()))
print(len(contains.keys()))

contains
```
